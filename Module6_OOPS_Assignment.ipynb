Answer: The five key concepts of Object-Oriented Programming (OOP) are:
 1 - Encapsulation: Encapsulation refers to bundling the data (attributes) and methods (functions) that operate on the data into a single unit, called a class.
 2- Inheritance: Inheritance allows one class (child or subclass) to inherit attributes and methods from another class (parent or superclass).
 3- Polymorphism: Polymorphism enables the ability to use the same interface or method name for different underlying data types or classes.
 4- Abstraction: Abstraction focuses on exposing only the essential details of an object while hiding its complex implementation.
 5- Composition: Composition involves building complex objects by combining other objects. It focuses on a "has-a" relationship instead of the "is-a" relationship provided by inheritance.

```python
#QSN6-  Describe the five types of inheritance in Python. Provide a simple example of multiple inheritance.?
'''
Ans: 1- Single Inheritance:A single child class inherits from a single parent class.
Example:
class Parent:
    def greet(self):
        print("Hello from Parent")

class Child(Parent):
    pass

obj = Child()
obj.greet()  # Output: Hello from Parent

2- Multiple Inheritance : A child class inherits from more than one parent class.
Example:
class Parent1:
    def greet1(self):
        print("Hello from Parent1")

class Parent2:
    def greet2(self):
        print("Hello from Parent2")

class Child(Parent1, Parent2):
    pass

obj = Child()
obj.greet1()  # Output: Hello from Parent1
obj.greet2()  # Output: Hello from Parent2

3- MultiLevel Inheritance: A child class inherits from a parent class, and another class inherits from the child Class.
Example:
class Grandparent:
    def greet(self):
        print("Hello from Grandparent")

class Parent(Grandparent):
    pass

class Child(Parent):
    pass

obj = Child()
obj.greet()  # Output: Hello from Grandparent

4- Hierarchical Inheritance: Multiple child classes inherit from the same parent class.
Example:
class Parent:
    def greet(self):
        print("Hello from Parent")

class Child1(Parent):
    pass

class Child2(Parent):
    pass

obj1 = Child1()
obj1.greet()  # Output: Hello from Parent
obj2 = Child2()
obj2.greet()  # Output: Hello from Parent

5- Hybrid Inheritance: A combination of two or more types of inheritance.
Example:
class Parent:
    def greet(self):
        print("Hello from Parent")

class Child1(Parent):
    pass

class Child2(Parent):
    pass

class GrandChild(Child1, Child2):
    pass

obj = GrandChild()
obj.greet()  # Output: Hello from Parent
'''
```

```python
#QSN7-  What is the Method Resolution Order (MRO) in Python? How can you retrieve it programmatically?
'''
Ans: Method Resolution Order (MRO) in Python: MRO determines the order in which Python looks for methods and
    attributes in a class hierarchy when there are multiple inheritance paths.MRO ensures that:
    * A consistent order is followed when searching for methods or attributes.
    * It resolves potential conflicts in multiple inheritance scenarios.

 How to Retrieve MRO Programmatically :
  We can retrieve the MRO of a class using:
  class MyClass:
    pass

print(MyClass.mro())

example:
class A:
    def show(self):
        print("A")

class B(A):
    def show(self):
        print("B")

class C(A):
    def show(self):
        print("C")

class D(B, C):
    pass

obj = D()
obj.show()  # Output: B

# Retrieve MRO
print(D.mro())  # Output: [<class '__main__.D'>, <class '__main__.B'>, <class '__main__.C'>, <class '__main__.A'>, <class 'object'>]
print(D.__mro__)  # Output: Same as above
'''
```

```python
#QSN8- Create an abstract base class `Shape` with an abstract method `area()`. Then create two subclasses
#`Circle` and `Rectangle` that implement the `area()` method.?
'''
Ans:
from abc import ABC, abstractmethod
import math

# Abstract Base Class
class Shape(ABC):
    @abstractmethod
    def area(self):
        """Abstract method to compute the area of the shape"""
        pass

# Subclass Circle
class Circle(Shape):
    def __init__(self, radius):
        self.radius = radius

    def area(self):
        """Calculates the area of a circle"""
        return math.pi * (self.radius ** 2)

# Subclass Rectangle
class Rectangle(Shape):
    def __init__(self, width, height):
        self.width = width
        self.height = height

    def area(self):
        """Calculates the area of a rectangle"""
        return self.width * self.height
circle = Circle(5)  # Circle with radius 5
rectangle = Rectangle(4, 6)  # Rectangle with width 4 and height 6

print(f"Circle area: {circle.area():.2f}")  # Output: Circle area: 78.54
print(f"Rectangle area: {rectangle.area()}")  # Output: Rectangle area: 24

Explanation:
Abstract Base Class Shape:Inherits from ABC (Abstract Base Class) provided by the abc module.
Subclass Circle: Defines an __init__ method to initialize the radius, Implements the area() method to calculate the area of a circle using the formula πr2.
Subclass Rectangle: Defines an __init__ method to initialize the width and height.
 and Implements the area() method to calculate the area of a rectangle using the formula width×height.
 '''
```

```python
#QSN9-  Demonstrate polymorphism by creating a function that can work with different shape objects to calculateand print their areas?
'''
Ans: Below are demonstrate polymorphism in Python by creating a function that calculates and prints the area of different shapes:

from abc import ABC, abstractmethod
import math

# Abstract Base Class
class Shape(ABC):
    @abstractmethod
    def area(self):
        pass

# Subclass Circle
class Circle(Shape):
    def __init__(self, radius):
        self.radius = radius

    def area(self):
        return math.pi * (self.radius ** 2)

# Subclass Rectangle
class Rectangle(Shape):
    def __init__(self, width, height):
        self.width = width
        self.height = height

    def area(self):
        return self.width * self.height

# Subclass Triangle
class Triangle(Shape):
    def __init__(self, base, height):
        self.base = base
        self.height = height

    def area(self):
        return 0.5 * self.base * self.height

# Polymorphic Function
def print_area(shape):
    print(f"The area of the {type(shape).__name__} is: {shape.area()}")
circle = Circle(5)        # Circle with radius 5
rectangle = Rectangle(4, 6)  # Rectangle with width 4 and height 6
triangle = Triangle(3, 8)    # Triangle with base 3 and height 8

# Using the polymorphic function
print_area(circle)        # Output: The area of the Circle is: 78.53981633974483
print_area(rectangle)     # Output: The area of the Rectangle is: 24
print_area(triangle)      # Output: The area of the Triangle is: 12.0
'''
```

```python
#QSN12-  Create a decorator that measures and prints the execution time of a function.?
'''
Ans: Below are example, how to  create a Python decorator that measures and prints the execution time of a function:

import time

# Decorator to measure execution time
def measure_time(func):
    def wrapper(*args, **kwargs):
        start_time = time.time()  # Record the start time
        result = func(*args, **kwargs)  # Call the original function
        end_time = time.time()  # Record the end time
        execution_time = end_time - start_time  # Calculate the time taken
        print(f"Execution time of {func.__name__}: {execution_time:.4f} seconds")
        return result
    return wrapper
 @measure_time
def slow_function():
    time.sleep(2)  # Simulate a slow function by sleeping for 2 seconds
    print("Function has completed.")

@measure_time
def fast_function():
    time.sleep(0.5)  # Simulate a faster function by sleeping for 0.5 seconds
    print("Function has completed.")

# Call the functions
slow_function()  # Output will include the execution time for slow_function
fast_function()  # Output will include the execution time for fast_function
'''
```

```python
#QSN13-  Explain the concept of the Diamond Problem in multiple inheritance. How does Python resolve it?
'''
Ans: The Diamond Problem is a complication that can occur in object-oriented programming when using
multiple inheritance. It involves a situation where a class inherits from two classes, and those classes
both inherit from a common base class. The issue arises when the derived class has to resolve which
path (method or attribute) to inherit from the common ancestor if both parent classes have their
own implementation of the same method or attribute.

Example:

class A:
    def hello(self):
        print("Hello from A")

class B(A):
    def hello(self):
        print("Hello from B")

class C(A):
    def hello(self):
        print("Hello from C")

class D(B, C):
    pass

d = D()
d.hello()  # Output: "Hello from B"
'''
```
